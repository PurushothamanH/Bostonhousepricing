```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Boston.csv')
df.head()
```

```python
df.rename(columns = {'Unnamed: 0':'ind_value'},inplace=True)
df.head()
```

```python
df.drop('ind_value',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
## to find correlation
se=df.corr()
se
```

```python
sns.pairplot(df.corr())
```

```python
plt.scatter(se["crim"],se["medv"])
plt.xlabel("crime rate")
plt.ylabel("Price")
```

```python
sns.regplot(data=se,x="crim",y="medv")
```

```python
## spliting the dataset into depedent and indepedent features...
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
## preprocessing...
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
## we use fit_transform only in train but in test we use only transform...
## to prevent data leakege....
x_train=sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
## modeling.....
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
## Predicted values....
y_predict = linear.predict(x_test)
y_predict
```

```python
## actualvalues....
y_test
```

```python
## the linear.coef povides that the coeficient values of each columns with depedent features....
## in simple -1.108 is coefficient of 1 feature crime with respct to medv...
print(linear.coef_)
```

```python
## provides intercept values...
print(linear.intercept_)
```

```python
## to find that the model works well using ploting...
## this show that the actual and predicted almost same.....align in same places....
plt.scatter(y_test,y_predict)
```

```python
## and another method is that the residuals must follow normal distribution...
residuals = y_test-y_predict
residuals
```

```python
sns.displot(residuals)
```

```python
sns.displot(residuals,kind="kde")
```

```python
## third method is to take predicted and residuals... if both are scattered uniformally then model is good...
## some  values are lie outside are the outliers..
plt.scatter(y_predict,residuals)
```

```python
 ## performance metrices...
from sklearn.metrics import r2_score
print(r2_score(y_test,y_predict))
```
